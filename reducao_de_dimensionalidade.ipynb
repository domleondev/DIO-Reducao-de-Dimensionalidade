```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Importação específica para Google Colab

# Carregar a imagem
img = cv2.imread("/content/Lenna_(test_image).png")


# Converter para escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar o Gaussian Blur
suave = cv2.GaussianBlur(gray, (7, 7), 0)

# Binarizar a imagem (preto e branco)
(T, bin) = cv2.threshold(suave, 160, 255, cv2.THRESH_BINARY)

# Redimensionar imagens para ter dimensões iguais
height, width = gray.shape
color_resized = cv2.resize(img, (width, height))

# Combinar as imagens para exibição
resultado = np.hstack([color_resized, cv2.cvtColor(suave, cv2.COLOR_GRAY2BGR), cv2.cvtColor(bin, cv2.COLOR_GRAY2BGR)])

# Exibir a imagem (compatível com Google Colab)
cv2_imshow(resultado)

# Para sistemas locais, use `cv2.imshow` e finalize com `cv2.waitKey` e `cv2.destroyAllWindows`
# cv2.imshow("Imagens: Original, Suavizada e Binarizada", resultado)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
